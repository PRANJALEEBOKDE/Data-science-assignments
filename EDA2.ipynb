## 2. Encoding Techniques:

```python


from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# One-Hot Encoding for categorical variables with less than 5 categories
onehot_encoder = OneHotEncoder()
onehot_encoded = onehot_encoder.fit_transform(data[['relationship', 'age']])

# Label Encoding for categorical variables with more than 5 categories
label_encoder = LabelEncoder()
data['label_encoded_column'] = label_encoder.fit_transform(data['occupation'])
data['label_encoded_column']
```
